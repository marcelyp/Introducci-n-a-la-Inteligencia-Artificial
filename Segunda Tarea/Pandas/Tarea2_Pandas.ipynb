```python
#Eliminar valores omitidos

fish.dropna(subset=["Species"], axis=0, inplace=False) #inplace:modifique directamente la BD
```

```python
#Manipular datos textuales
fish["Species"] = (fish["Species"]
                            #.str.lower() #Pone en minustula todo el texto
                            #.str.replace("[^a-záéíóúüñ ]","") # Reemplazar valores por otros
                            #.str.replace(" +"," ") 
                            #.str.strip() #Elimina espacios al principio y al final de cada cadena
                            .str.title() #Primer letra de cada palabra en mayúscula
                           ) 
fish["Species"]
```

```python
#Creación de columnas a partir de otras

fish["ph"]= fish["Weight"]/fish["Height"]
print(fish[["ph","Weight","Height"]])
```

```python
#Renombrar columnas
print(fish.columns)
fish.rename(columns={"ph":"ph_"},inplace=True) #renombrar columna
print(fish.columns) #Verificación
```

```python
#Eliminar filas y columnas
#Filas
print(fish.shape)
fish = fish.drop(fish.iloc[0:3,].index)
fish = fish.drop(fish[fish['Species']=='Parkki'].index)
print(fish.shape)

#Columnas
fish.drop(['Length3'], axis=1,inplace=True) #Eliminar columna de la BD
fish.columns
```

```python
#Conteo / sumar de acuerdo a una característica
fish.groupby('Species')[['Weight', 'Length1']].sum()
```

```python
#Organizar los datos por Weight
fish.sort_values(["Weight"], ascending=False)
```

```python
#Aplicar ambos métodos al mismo tiempo
fish2=(fish.groupby('Species')[['Weight', 'Length1']]
       .sum()
       .sort_values(["Weight"], ascending=True))
display(fish2)
```
